```python
data.columns = col_names
```

```python
target = data["nominal"]
train = data.drop("nominal", axis=1)
```

```python
scaler = preprocessing.StandardScaler()
scaler.fit(train)

train = scaler.transform(train)
```

```python
len(train)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf = clf.fit(train, target)
logic_regression = linear_model.LogisticRegression()
logic_regression = linear_regression.fit(train, target)
```

```python
def feature_importances_name(features_array):
    return [(features_array[i], col_names[i]) for i in range(len(features_array))]
```

```python
features_tree = feature_importances_name(clf.feature_importances_)
features_tree.sort(reverse = True)
```

```python
features_tree
```

```python
test = pd.read_csv("testing.csv")
```

```python
test.columns = col_names
```

```python
expecting = test["nominal"]
test = test.drop("nominal", axis=1)
cols=test.columns
test = preprocessing.normalize(test)
```

```python
data_outputs = clf.predict(test)
```

```python
def count_procent(data_outputs, expect_values):
    right_count = 0
    
    for ind in range(len(data_outputs)):
        if data_outputs[ind] == expect_values[ind]:
            right_count += 1
    return (right_count / len(data_outputs)) * 100
```

```python
importances = clf.feature_importances_
std = np.std([tree.feature_importances_ for tree in clf.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(test.shape[1]):
    print("%s. feature %d %s (%f)" % (f + 1, indices[f], cols[indices[f]], importances[indices[f]]))

# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(test.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(test.shape[1]), indices)
plt.xlim([-1, test.shape[1]])
plt.show()
```

```python
selector = RFE(logic_regression, n_features_to_select=1)
selector.fit(train, target)
order = selector.ranking_
order
```
