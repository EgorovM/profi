```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn import tree, linear_model
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
data = pd.read_csv("gender_task_set_open.csv")
```

```python
ids = data["id"]
```

```python
data = data[
    [
      'videos_count', 
      'photos_count', 
      'gifts_count', 
      'audios_count', 
      'friends_count', 
      'followers_count'
    ]
]
```

```python
train_data = pd.read_csv("gender.csv")
```

```python
train_data
```

```python
target = train_data["SEXES"]
train = train_data.drop("SEXES", axis=1)
train = train.drop("IDS", axis=1)
```

```python
scaler = preprocessing.StandardScaler()
scaler.fit(train)

train = scaler.transform(train)
data  = scaler.transform(data)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100)
clf.fit(train, target)
```

```python
data_outputs = clf.predict(data)
```

```python
ans = ids.to_frame()
ans["sex"] = data_outputs
ans.to_csv("output.csv", index = False)
```
